### 6.1 Handling unwanted or redundant columns

```python
# We will drop this columns which does not give any value adding in the currently depth of analysis.
data.drop(labels='engV', axis=1,inplace=True)
data
```

<a name = Section62></a>
### 6.2 Handling missing data

```python
#missing data
total = data.isnull().sum().sort_values(ascending=False)
percent = ((data.isnull().sum()/data.isnull().count())*100).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
#missing_data.head(20)
print(missing_data)
```

```python
data['drive'].describe()
```

**Drive column has 511 missing data. these 'nan' values can be replaced with mode of that column (i.e front) which has a higher frequency of occurance (5188/9065*100 = 57.23%). Here in this case replacing with front we have nearly 57% of chance being correct.**

```python
data['drive'] = data['drive'].fillna(data['drive'].mode()[0])
```

```python
data['drive'].isnull().sum()
```

<a name = Section63></a>
### 6.3 Handling Duplicates in the Data

```python
data.drop_duplicates(inplace=True) # we can drop the duplicate rows 
data
```

<a name = Section64></a>
### 6.4 Handling redundant data 
(Price and mileage cannot be zero, so these data has been drop as of now these may be redundancy while data collection)

```python
price_zero_idx = data[data['price']==0].index.tolist()
data.drop(axis=0, index=price_zero_idx, inplace=True)
```

```python
mileage_zero_idx = data[data['mileage']==0].index.tolist()
data.drop(axis=0, index=mileage_zero_idx, inplace=True)
```

```python
data
```

<a name = Section65></a>
### 6.5 Handing the Price Outliers 

```python
print("On an average a car sale price is {:.4f} $, 99% of car price is {:.0f} or less, while the maximum price ever recorded is {:.0f}.".format(data['price'].mean(),data['price'].quantile(0.99), data['price'].max()))
```

**This can be seen on the Graph Below** 

```python
data.groupby('car').price.mean().plot(kind='box',x='price')
```

**Removing the outliers from the price, that is we drop the rows of price where its more than 95000.**

```python
Higher_end_cars_df = data[data['price']>=95000]
Higher_end_cars_lst = data[data['price']>=95000].index.tolist()
data.drop(axis=0, index = Higher_end_cars_lst, inplace =True)
data
```

```python
data.groupby('car').price.mean().plot(kind='box',x='price')
```

<a name = Section66></a>
### 6.6 Handing the Mileage Outliers 

```python
print("On an average a Mileage is {:.2f} mgl, 99% of car's Mileage is {:.0f} or less, while the maximum Mileage ever recorded is {:.0f}.".format(data['mileage'].mean(),data['mileage'].quantile(0.99), data['mileage'].max()))
```

**This can be seen on the Graph Below** 

```python
data.mileage.plot(kind='box')
```

**Removing the outliers from the mileage, that is we drop the rows of cars where mileage is more than 420.**

```python
Higher_end_cars_df = data[data['mileage']>=420]
Higher_end_cars_lst = data[data['mileage']>=420].index.tolist()
data.drop(axis=0, index = Higher_end_cars_lst, inplace =True)
data
```
